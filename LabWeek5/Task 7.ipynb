### Q.No 1 Create a 1D array of 12 elements and reshape it into a 3×4 matrix.


```python
import numpy as np
```

```python
array = np.arange(start=0 , stop=12, step= 1)
print(array)
```

```python
reshape_array = array.reshape(3,4)
print(reshape_array)
```

### Q.No 2 Reshape a 6-element 1D array into a 2×3 matrix, then flatten it back to 1D using flatten().


```python
array = np.random.randint(low=1, high= 7, size=6)
print(array)
```

```python
reshape_array = array.reshape((2,3))
print(reshape_array)
```

```python
flattened_array = reshape_array.flatten()
print(flattened_array)
```

### Q.No 3 Transpose a 3×4 random array (swap rows and columns).


```python
random_array = np.random.randint(low=10, high=30, size=(3,4))
print(random_array)
```

```python
transpose_array = random_array.transpose()
print(transpose_array)
```

### Q.No 4 Reshape a 2×2×3 array into a 3×4 matrix using reshape.


```python
array = np.linspace(start=10, stop= 30, num=12).reshape(2,2,3)
print(array)
```

```python
reshape_array = array.reshape(3,4)
print(reshape_array)
```

### Q.No 5 Create a 4×4 identity matrix and flatten it in row-major order (C order).


```python
identity_matrix = np.eye(4)
print(identity_matrix)
```

```python
flattened_array = identity_matrix.flatten(order="C")  # row wise flatten  .  if "F" then column wise flatten
print(flattened_array)
```
