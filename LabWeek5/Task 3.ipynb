### Q.No 1 Create a 1D array [10, 20, 30, 40, 50, 60, 70] and extract the element at index 4.


```python
import numpy as np
```

```python
array_1d = np.array([10,20,30,40,50,60, 70])
```

```python
print(f"Value at index 4 is {array_1d[4]}")
```

### Q.No 2 Using the same array, slice and get elements from index 2 to 5 (inclusive of 2, exclusive of 6).


```python
array_1d[2 : 6]  #  inluded : excluded
```

### Q.No 3 Create a 3×4 array of random integers (0–9) and extract the entire second row.

```python
random_array = np.random.randint(9, size=(3,4))
random_array
```

```python
print(f"Second row of random_array is {random_array[1]}")
```

### Q.No 4 From the same 3×4 array, extract the sub-matrix consisting of the first two rows and last two columns.

```python
random_array[0:2 , 2:4]   # after : not included
```

```python
random_array[ : 2, 2: ]
```
