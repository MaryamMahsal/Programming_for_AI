### Q.No 3 Create a 1D array [1, 2, 3] and add it to a 3×3 matrix of zeros using broadcasting.


```python
array_1d = np.array([1,2,3])
array_1d
```

```python
zeros_matrix = np.zeros(shape=(3,3))
zeros_matrix
```

```python
result = zeros_matrix + array_1d  # result is broadcasted matrix i.e [[1,2,4],[1,2,3],[1,2,3]] and then added to 3x3 zeros
result
```
