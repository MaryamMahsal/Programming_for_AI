## Q.No 1: Create a 1D array from the Python list [10, 20, 30, 40, 50].

```python
python_list = [10, 20, 30, 40, 50]
print(f"Pyhton List = { python_list}")
```

## Q.No 2: Create a 2D array (3 rows × 4 columns) filled with zeros.

```python
array_2d = np.zeros((3,4))
array_2d
```

```python
print("Number of Dimensions:", array_2d.ndim) # Number of dimension

# print(f"Number of Dimensions: {array_2d.ndim}:")
```

# Q.No 3: Create a 3D array of shape (2, 3, 4) filled with ones.

```python
array_3d = np.ones((2, 3, 4))  # First Blocks, Rows , Columns  , It can also takes dtype=int,float
print("3-D Array of Ones:")
print(array_3d)
```

```python
print(f"Number of  Dimensiona : { array_3d.ndim}")
```

## Q.No 4: Create a 1D array containing all integers from 0 to 19 using arange.


```python
#                        start(include) , stop(not included-> excluded) , step(increment in the presvious value to get the next value)
arange_arr_1d = np.arange(start= 0 , stop= 19, step= 2)   # or   np.arange(0,19,2)

print(arange_arr_1d)
```
