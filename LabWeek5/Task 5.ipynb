```python
import numpy as np
```

### Q.No 1 Compute the square root of every element in a 1D array [1, 4, 9, 16, 25].


```python
array_1d = np.array([1, 4, 9, 16, 25])
```

```python
sqrt_arr_1d = np.sqrt(array_1d)
```

```python
print(f"1-D array {array_1d} and it's square root is {sqrt_arr_1d}")
```

### Q.No 2 Compute the sine of every element in np.linspace(0, np.pi, 5).


```python
array = np.linspace(start=0, stop=np.pi, num=5)   # np.linspace(0, np.pi, 5) # start and stop both are included
array
```

```python
sine_array = np.sin(array)
sine_array
```

### Q.No 3 Calculate the exponential (exp) of a 2×3 array of random numbers between 0 and 2.


```python
nor_array = np.random.uniform(low=0, high=2, size=(2,3))     # (0, 2, size=(2,3)) or  (0, 2, (2,3))  start 0 is include but end 2 is excluded
nor_array
```

```python
exp_array = np.exp(nor_array)

exp_array
```

### Q.No 4 Compute the natural logarithm (log) of [1, 10, 100, 1000].

```python
array = np.array([1, 10, 100, 1000])
array
```

```python
nlog_array = np.log(array)  # natural log base e(2.17) , common log10 and binary log have base 2 i.e log2
```

```python
print(f"Arary :\n {array}")
print(f"Natural log array : \n {nlog_array}")
```
