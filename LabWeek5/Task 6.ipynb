### Q.No 1 Compute the total sum of all elements.


```python
import numpy as np

data = np.array([
                [10, 20, 30],
                [40, 50, 60],
                [70, 80, 90]
                ]
                )
data
```

### Q.No 2 Compute the mean of each column.


```python
mean_array = np.mean(data, axis=0)  # if axis= 1 then it caclulate mean by row-wise
mean_array
```

```python

# Manually
column_1 = np.mean(data[:, 0])
column_2 = np.mean(data[:, 1])
column_3 = np.mean(data[:,2])

column_1, column_2, column_3
```

### Q.No 3 Find the maximum value in the entire array.

```python
max_value = np.max(data)
print(f"Maximum value in the entire array is {max_value}" )
```

### Q.No 4 Compute the standard deviation of each row.

```python
np.std(data, axis=1)  # if axis = 0 then it did column wise
```

### Q.No 5 Find the minimum value in each column.


```python
np.min(data, axis=0)  # if axis= 1 then it will take smallest value from each row and return as 1-d array
```
