### Q.No 2 Calculate the average and standard deviation (std) of the column factor_1  and display the result.
### - •Pandas mean() and std()

```python
df['factor_1']
```

#### **Mean of Factor_1**

```python
mean_factor_1 = df['factor_1'].mean()
mean_factor_1 = np.round(mean_factor_1, 2) # round it two decimal places
```

```python
print(f"Mean of factor_1 is : {mean_factor_1}")
```

#### **Standard Deviation of Factor_1**

```python
std_factor_1 = df['factor_1'].std()
std_factor_1 = np.round(std_factor_1, 2)
```

```python
print(f"Standard Deviation of factor_1 is : {std_factor_1}")
```
