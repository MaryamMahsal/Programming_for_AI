### Apply a lambda function to calculate the square root of each value in the data frame import math  math.sqrt()


```python
import numpy as np
import pandas as pd
from math import sqrt
```

```python
array = np.random.randint(low=1, high =100, size=(3,5))
array
```

```python
df = pd.DataFrame(data=array, columns=[1,2,3,4,5], index=['a', 'b', 'c'])
```

### **Lambda Function**

#### *Implemenation of Lambda Function*

```python
df.map(lambda cell_value: sqrt(cell_value))
```
