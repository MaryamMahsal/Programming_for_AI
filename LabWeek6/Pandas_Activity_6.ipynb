### Q.No 1 Create a numpyarray of the shape (3,5): 3 rows and 5 columns  with random values from 1 to 100.

###    - Use np.random.randint(low,high, (shape))

```python
import numpy as np
import pandas as pd
```

```python
array = np.random.randint(low=1, high=100, size=(3,5))
print("Numpy array having 3 Row ad 5 columns")
print(array)
```

### Q.No 2 Use the numpyarray to generate a Data Frame with index = a,b,c and columns 1,2,3,4, 5•

```python
df = pd.DataFrame(
    data = array, 
    columns = [1,2,3,4,5], 
    index = ['a', 'b', 'c']  # if no of rows != no of indices then error will occure.
    )  
```
